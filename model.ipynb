```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.svm import SVC
import pandas as pd
```

```python
data = pd.read_csv("data/csv/reports-data.csv")
```

```python
data = data.drop([
'releaseDate',
'pibData4Weeks',
'pibData1Week',
'pibDataToday',
'igpData4Weeks',
'igpData1Week',
'igpDataToday',
'investimentoData4Weeks',
'cambioData4Weeks',
'cambioData1Week',
'cambioDataToday',
'ipcaAdminData1Week',
'ipcaAdminDataToday',
'contaCorrenteData4Weeks',
'balComData4Weeks',
'balComDataToday',
'investimentoData1Week',
'investimentoDataToday',
'dividaLiquidaData4Weeks',
'dividaLiquidaData1Week',
'dividaLiquidaDataToday',
'resPrimarioData4Weeks',
'resPrimarioData1Week',
'resNominalData4Weeks',
'resNominalData1Week'
], axis=1)

data.head()
```

```python
features = data.drop("stockDirection", axis=1)
target = data["stockDirection"]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.8)
```

```python

def create_models():
    models = [
        ('RandomForestClassifier', RandomForestClassifier(n_estimators=100, max_depth=5)),
        ('SVC', SVC(kernel='linear')),
        ('KNeighborsClassifier', KNeighborsClassifier(n_neighbors=4)),
        ('LogisticRegression', LogisticRegression())
    ]
    return models
```

```python
def train_and_evaluate(models, X_train, X_test, y_train, y_test):
    
    results = []
    
    for name, model in models:
        # training
        print(f"Training the model {name} with data from training set...")
        model.fit(X_train, y_train)

        # predictions with data from test set
        y_pred = model.predict(X_test)

        # calculation of metrics
        report = classification_report(y_test, y_pred)
        print("Classification Report\n", report)

        results.append([model, {'model': name, 'predictions': y_pred, 'report': report,}])           

    return results
```

```python
models = create_models()
models
```

```python
results = train_and_evaluate(models, X_train, X_test, y_train, y_test)
results
```
